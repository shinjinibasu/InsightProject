```python
import pandas as pd
import numpy as np
```

```python
users = pd.read_csv('goodreads2/user_w_uid_bavg.csv')
books = pd.read_csv('goodreads2/all_book_div.csv',index_col=0)
```

```python
id_title = pd.Series(books.original_title.values,index=books.book_id).drop_duplicates().to_dict()
```

```python
users['title'] = users['goodreads_book_id'].map(id_title)
```
